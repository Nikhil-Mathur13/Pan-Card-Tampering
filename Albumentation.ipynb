```python
# augmentor = alb.Compose([alb.RandomCrop(width = 620, height = 620),
#                          alb.HorizontalFlip(p=0.5),
#                          alb.RandomBrightnessContrast(p = 0.2),
#                          alb.RandomGamma(p=0.2) , alb.RGBShift(p=0.2),
#                          alb.VerticalFlip(p=0.5)],
#                          bbox_params =  alb.BboxParams(format = 'albumentations' ,
#                                                  label_fields = ['class_labels']))

# class_labels = {'RandomCrop'  , 'HorizontalFlip'  , 'RandomBrightnessContrast'  , 'RGBShift'  , 'VerticalFlip' , 'MedianBlur' , 'HueSaturationValue'}


# augmentor = alb.Compose([alb.RandomCrop(width = 620, height = 620),
#                          alb.HorizontalFlip(p=0.2),
#                          alb.RandomBrightnessContrast(p = 0.5),
#                          alb.RandomGamma(p=0.5) , alb.RGBShift(p=0.5),
#                          alb.VerticalFlip(p=0.2),
#                          alb.MedianBlur(p=0.2),
#                          alb.HueSaturationValue(p=0.5)],
#                          alb.BboxParams(format = 'albumentations' ,
#                                                  label_fields = ['class_labels'])
#                         )


# Define the parameter list
params = [
    {"transform": alb.RandomCrop, "params": {"width": 620, "height": 620}},
    {"transform": alb.HorizontalFlip, "params": {"p": 0.2}},
    {"transform": alb.RandomBrightnessContrast, "params": {"p": 0.5}},
    {"transform": alb.RandomGamma, "params": {"p": 0.5}},
    {"transform": alb.RGBShift, "params": {"p": 0.5}},
    {"transform": alb.VerticalFlip, "params": {"p": 0.2}},
    {"transform": alb.MedianBlur, "params": {"p": 0.2}},
    {"transform": alb.HueSaturationValue, "params": {"p": 0.5}},
    {"transform": alb.ShiftScaleRotate, "params": {"shift_limit": 0.0625, "scale_limit": 0.2, "rotate_limit": 15, "p": 0.5}},
    {"transform":alb.ChannelShuffle , "params":{"p":0.5}},
    {"transform": alb.ToGray,"params": {"p":0.2}}

]

# Create the augmentor with the parameter list
augmentor = alb.Compose([t["transform"](**t["params"]) for t in params],
                         bbox_params=alb.BboxParams(format='albumentations'))


import cv2
image=cv2.imread('/content/Photo.jpg')
image=cv2.cvtColor(image,cv2.COLOR_BGR2RGB)

for i in range(10):
  bboxes = []
  transformed = augmentor(image=image ,bboxes=bboxes)
  transformed_image = transformed["image"]
  plt.imshow(transformed_image)
  plt.show()

```
